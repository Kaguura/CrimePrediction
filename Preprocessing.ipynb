```python
import pandas as pd

incidents = pd.read_csv("Data/Removed_Incident.csv")
incidents = incidents.loc[:, ~incidents.columns.str.contains('^Unnamed')]
print(incidents.head(10))
```

```python
from sklearn.preprocessing import LabelEncoder

category_le = LabelEncoder()
incidents["Category"] = category_le.fit_transform(incidents["Category"].astype("str"))
category_classes = category_le.classes_

dayofweek_le = LabelEncoder()
incidents["DayOfWeek"] = dayofweek_le.fit_transform(incidents["DayOfWeek"].astype("str"))
dayofweek_classes = dayofweek_le.classes_

incidents['Time'] = [int(inc[:2])*60 + int(inc[3:]) for inc in incidents['Time']]

pdDistrict_le = LabelEncoder()
incidents["PdDistrict"] = pdDistrict_le.fit_transform(incidents["PdDistrict"].astype("str"))
pdDistrict_classes = pdDistrict_le.classes_

resolution_le = LabelEncoder()
incidents["Resolution"] = resolution_le.fit_transform(incidents["Resolution"].astype("str"))
resolution_classes = resolution_le.classes_
```
